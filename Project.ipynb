```python
import os
import numpy as np
import matplotlib.pylab as plt
from sklearn.ensemble import RandomForestClassifier
from astropy.table import Table
```

```python
data = Table.read('test_catalog.fits')
data
```

```python
modelmag = data['MODELMAG']+data['EXTINCTION'] # [u,g,r,i,z] array for each object and correction for extinction (in [u,g,r,i,z])
w1mpro = data['W1MPRO']
r_w1 = modelmag[:,2]-w1mpro[:]
r_i = modelmag[:,2]-modelmag[:,3]
goodz = data['GOODZ']
redshift = data['Z']
U_B = data['UB_0']
# discriminate good LRGs — objects with GOODZ=1, Z > 0.6, and U-B > 1
y = np.logical_and(goodz==1,redshift>0.6,U_B>1).astype('int')
# bad = np.logical_not(np.logical_and(goodz==1,redshift>0.6,U_B>1))
r_w1_good = r_w1[y==1]
r_w1_bad = r_w1[y==0]
r_i_good = r_i[y==1]
r_i_bad = r_i[y==0]
print r_w1.shape
print r_w1_good.shape
print r_w1_bad.shape
```

```python
plt.figure(1)
plt.scatter(r_i_good,r_w1_good,s=1,color='r')
plt.scatter(r_i_bad,r_w1_bad,s=1,color='b')
plt.xlim(0.5,4.2)
plt.ylim(2,14)
plt.xlabel('r-i')
plt.ylabel('r-w1')
plt.figure(2)
plt.scatter(r_i_good,r_w1_good,s=1,color='r')
plt.xlim(0.5,4.2)
plt.ylim(2,14)
plt.xlabel('r-i')
plt.ylabel('r-w1')
plt.figure(3)
plt.scatter(r_i_bad,r_w1_bad,s=1,color='b')
plt.xlim(0.5,4.2)
plt.ylim(2,14)
plt.xlabel('r-i')
plt.ylabel('r-w1')
```

```python
X = np.column_stack((r_i,r_w1))
clf = RandomForestClassifier(n_estimators=15)
clf.fit(X,y)
print clf.score(X,y)

plot_step = 0.02

x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),np.arange(y_min, y_max, plot_step))

Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
cs = plt.contourf(xx, yy, Z, cmap=plt.cm.jet)
plt.scatter(r_i_good,r_w1_good,s=1,color='r')
plt.scatter(r_i_bad,r_w1_bad,s=1,color='b')
plt.xlim(0.5,4.2)
plt.ylim(2,14)
plt.xlabel('r-i')
plt.ylabel('r-w1')
plt.axis("tight")
```
